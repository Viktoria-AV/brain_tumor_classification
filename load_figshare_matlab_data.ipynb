```python
#loads data from the figshare zip .mat files
#they have to be unzipped first after downloaded

import numpy as np
import h5py
import cv2

import glob
d=1
lab=[]
mask= []
im= []
PID=[] 

for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/tumor_data_1/*.mat'):
    f = h5py.File(filename,'r') 
    data = f.get('cjdata')
    im1=(data.get('image'))
    im.append(np.array(im1))

    mask1=data.get('tumorMask')
    mask.append(np.array(mask1))

    PID1=data.get('PID')
    PID.append(np.array(PID1))
    #print("PID1:",PID1 )

    lab1=data.get('label')
    lab.append(np.array(lab1))
    d=d+1


label_part1=np.array(lab)
label_part1=np.squeeze(label_part1)

PID_part1=np.array(PID)
PID_part1=np.squeeze(PID_part1)

mask_part1=np.array(mask)
mask_part1=np.squeeze(mask_part1)

im_part1=np.array(im)
im_part1=np.squeeze(im_part1)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/label_part1.npy',label_part1)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/mask_part1.npy',mask_part1)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/im_part1.npy',im_part1)
np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/PID_part1.npy',PID_part1)

##################################
################################## part 2 
d=767

lab=[]
mask= []
im= []
PID=[] 

for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/tumor_data_2/*.mat'):
    f = h5py.File(filename,'r') 
    data = f.get('cjdata')
    im1=(data.get('image'))
    im.append(np.array(im1))

    mask1=data.get('tumorMask')
    mask.append(np.array(mask1))

    PID1=data.get('PID')
    PID.append(np.array(PID1))
    #print("PID1:",PID1 )

    lab1=data.get('label')
    lab.append(np.array(lab1))
    d=d+1


label_part2=np.array(lab)
label_part2=np.squeeze(label_part2)

PID_part2=np.array(PID)
PID_part2=np.squeeze(PID_part2)

mask_part2=np.array(mask)
mask_part2=np.squeeze(mask_part2)

im_part2=np.array(im)
im_part2=np.squeeze(im_part2)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/label_part2.npy',label_part2)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/mask_part2.npy',mask_part2)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/im_part2.npy',im_part2)
np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/PID_part2.npy',PID_part2)

####part 3
d=1533


lab=[]
mask= []
im= []
PID=[] 

for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/tumor_data_3/*.mat'):
    f = h5py.File(filename,'r') 
    data = f.get('cjdata')
    im1=(data.get('image'))
    im.append(np.array(im1))

    mask1=data.get('tumorMask')
    mask.append(np.array(mask1))

    PID1=data.get('PID')
    PID.append(np.array(PID1))

    lab1=data.get('label')
    lab.append(np.array(lab1))
    d=d+1


label_part3=np.array(lab)
label_part3=np.squeeze(label_part3)

PID_part3=np.array(PID)
PID_part3=np.squeeze(PID_part3)

mask_part3=np.array(mask)
mask_part3=np.squeeze(mask_part3)

im_part3=np.array(im)
im_part3=np.squeeze(im_part3)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/label_part3.npy',label_part3)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/mask_part3.npy',mask_part3)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/im_part3.npy',im_part3)
np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/PID_part3.npy',PID_part3)



d=2299
lab=[]
mask= []
im= []
PID=[] 

for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/tumor_data_4/*.mat'):
    f = h5py.File(filename,'r') 
    data = f.get('cjdata')
    im1=(data.get('image'))
    im.append(np.array(im1))

    mask1=data.get('tumorMask')
    mask.append(np.array(mask1))

    PID1=data.get('PID')
    PID.append(np.array(PID1))
    #print("PID1:",PID1 )

    lab1=data.get('label')
    lab.append(np.array(lab1))
    d=d+1


label_part4=np.array(lab)
label_part4=np.squeeze(label_part4)

PID_part4=np.array(PID)
PID_part4=np.squeeze(PID_part4)

mask_part4=np.array(mask)
mask_part4=np.squeeze(mask_part4)

im_part4=np.array(im)
im_part4=np.squeeze(im_part4)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/label_part4.npy',label_part4)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/mask_part4.npy',mask_part4)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/im_part4.npy',im_part4)
np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/labels/PID_part4.npy',PID_part4)


#########
#error in part 2 images and masks- make aarrays 3 dimensional, its packed in 0 dimensions now (only for part2)
image_size=512
#resize images, to sg that can still be fitted to easily from computational perspective
def resize_images(images, image_size):
  images_new=[]
  for i in range (0,len(images)):
    #print(i)
    image_new = cv2.resize(images[i], dsize=(image_size,image_size), interpolation=cv2.INTER_CUBIC)
    images_new.append(image_new)
    #smaller size so it dont crash
  images= np.array(images_new)
  return images
mask_part2= resize_images(mask_part2, image_size)
im_part2= resize_images(im_part2, image_size)



#merge 4 subfolders
tumor_label=np.concatenate((label_part1,label_part2,
                            label_part3,label_part4),axis=0)


tumor_PIDs=np.concatenate((PID_part1,PID_part2,
                            PID_part3,PID_part4),axis=0)

tumor_mask=np.concatenate((mask_part1,mask_part2,
                            mask_part3,mask_part4),axis=0)

tumor_im=np.concatenate((im_part1,im_part2,
                            im_part3,im_part4),axis=0)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/data_whole_official/tumor_label.npy',tumor_label)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/data_whole_official/tumor_PIDs.npy',tumor_PIDs)

np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/data_whole_official/tumor_mask.npy',tumor_mask)
np.save('/content/drive/MyDrive/Colab Notebooks/tumor_data_all/data_whole_official/tumor_im.npy',tumor_im)

 
```
